```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(seed=42)
X = rng.uniform(low=0,high=2,size=100)
y = 4 + 3 * X + rng.standard_normal(size=(100,1)) # true model: y = 4 + 3x + noise
```

```python
X_b = np.c_[np.ones((100,1)),X]# Add bias term (intercept)
```

```python
# GRADIEND DESCENT parameters
m = len(y)
n_iter = 100
learning_rate = 0.1
```

```python
loss_history = []
tetha = rng.standard_normal(size=(2,1))
```

```python
for i in range(n_iter):
    gradients = 2/m * X_b.T.dot(X_b.dot(tetha)-y)
    tetha = tetha - learning_rate * gradients
    loss = np.mean((X_b.dot(tetha)-y)**2) # MSE
    loss_history.append(loss)
    
```

```python
# Plot loss per iteration
plt.plot(range(n_iter), loss_history, label="Loss (MSE)")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss per Iteration during Gradient Descent")
plt.legend()
plt.grid(True)
plt.show()
```
