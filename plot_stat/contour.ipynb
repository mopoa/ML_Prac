```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
import numpy as np

# Example data
X = np.linspace(0, 1, 500)
Y = np.linspace(0, 1, 500)
# Y = np.random.randn(50)
X, Y = np.meshgrid(X, Y)
# Z = np.sin(np.sqrt(X**2 + Y**2))
Z = X+Y

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')  # instead of fig.gca()

s = ax.plot_surface(X, Y, Z, cmap=plt.cm.rainbow)
cset = ax.contour(X,Y,Z)
fig.colorbar(s, shrink=0.5)

plt.show()
```

```python
x = np.arange(-10,10,0.5)
y = np.arange(-10,10,0.5)
x,y = np.meshgrid(x,y)
z = x**2 + y**2
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection='3d')
s = ax.plot_surface(x,y,z,cmap=plt.cm.coolwarm)
c = ax.contour(x,y,z)
plt.colorbar(s,shrink=0.5,aspect=5)
plt.show()
```
