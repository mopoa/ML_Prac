```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.array([-20,-3,-2,-1,0,1,2,3,4,5,8,9,10,11,12,13,15,16,19,40,50])
```

```python
q1 = np.quantile(data,0.25,interpolation='lower')
print(q1)
```

```python
q2 = np.quantile(data,0.5,interpolation='lower')
print(q2)
```

```python
q3 = np.quantile(data,0.75,interpolation='lower')
print(q3)
```

```python
iqr = q3 - q1
iqr
```

```python
lv = q1 - 1.5 * iqr
```

```python
hv = q3 + 1.5 * iqr
```

```python
print(lv)
print(hv)
```
