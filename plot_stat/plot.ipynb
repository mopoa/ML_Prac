```python
import matplotlib.pyplot as plt
import numpy as np


x = list(range(1,6))
y = list(range(2,11,2))
print(x)
print(y)
```

```python
plt.plot(x,y,'g--p',label='y=2x')
plt.legend()
plt.title('test')
plt.xlabel('x')
plt.ylabel('y')
```

```python
x1 = x.copy()
y1 = [1,4,9,16,25]
```

```python
plt.plot(x1,y1,'b:D',label='y=x^2')
plt.legend()
```

```python
plt.plot(x,y,'r--p',label='y=2x')
plt.plot(x1,y1,'b:D',label='y=x^2')
```

```python
plt.subplot(121)
plt.plot(x,y,'r--P',label='y=2x')
plt.subplot(122)
plt.plot(x1,y1,'b:D',label= 'y=x^2')
plt.tight_layout()
plt.savefig('./plot.png')
```

```python
x = np.linspace(1,10,100)

plt.plot(x,np.sin(x),'r',marker='.',label='sine wave')
plt.legend()
plt.title('sin wave')
```
